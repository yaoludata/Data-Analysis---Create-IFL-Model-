#### IFL model:
#### I: interaction_rate represents the average rate of interaction per video
#### F: Frequency represents the average release cycle per video
#### L: love_rate represents the average like rate of the video

### Step 1 Import modules, read files, and filter data

#### 1.1 Import the module and read the file

```python
# import pandas

import pandas as pd
```

```python
# import dataset

bibli = pd.read_csv('bilibili.csv')

bibli
```

#### 1.2 Statistics

```python
# select the data which from '科学科普'

bibli = bibli[bibli['分区'] == '科学科普']
bibli
```

```python
# Use the groupby() and count() functions
# The data variable is grouped and counted by "author" and the result is assigned to the count variable

titlecount = bibli.groupby(bibli['author']).count()
```

```python
# Use the column index to select the column data for "title" in titlecount

titlecount = titlecount['title'].reset_index()
titlecount
```

```python
# Name titlecount's column index "author" and "times"

titlecount.columns = ["author","times"]
titlecount
```

```python
# Use the pd.merge() function to connect within the titlecount and bibli dataset are merged horizontally and assigned to bibli

bibli = pd.merge(bibli, titlecount, on = 'author', how = 'inner')
bibli
```

```python
# Using the Boolean index, the authors with frequencies greater than or equal to 5 is filtered out and assigned to bibli

bibli = bibli[bibli['times'] >= 5]

bibli
```

### Step 2 Create IFL model

#### 2.1 Calculate I

I: interaction_rate represents the average rate of interaction per video

I = (total number of danmu and total replies) / Total views / total videos x 100

```python
SumDanmu = bibli['danmu'].groupby(bibli['author']).sum()
SumDanmu
```

```python
SumReply = bibli['reply'].groupby(bibli['author']).sum()
SumReply
```

```python
SumView = bibli['view'].groupby(bibli['author']).sum()
SumView
```

```python
SumTitle = bibli['title'].groupby(bibli['author']).count()
SumTitle
```

```python
i = (SumDanmu+SumReply) / SumView / SumTitle * 100
i = i.reset_index()
i.columns = ['author', 'I']
i
```

#### 2.1 Caculate F
F: Frequency represents the average release cycle per video
F = 1/ (the interval between the last video released by the author and the date of the first video released by the author + 1) / The total number of videos * 100

```python
# check the datatype of date column

bibli.info()
```

```python
# transfer the date column from object to datetime

bibli['date'] = pd.to_datetime(bibli['date'])
```

```python
bibli.info()
```

```python
# Gets when the author updated the video earliest

DayEarly = bibli['date'].groupby(bibli['author']).min()
DayEarly
```

```python
# Get the date when the author last updated the video

DayLate = bibli['date'].groupby(bibli['author']).max()
DayLate
```

```python
# caculate the interval between the last video released by the author and the date of the first video released by the author

DayGap = (DayLate - DayEarly).dt.days
DayGap
```

```python
f = 1 / (DayGap + 1) / SumTitle * 100
f = f.reset_index()
f.columns = ['author', 'F']
```

#### 2.3 Caculate L
L represents the average like rate of the video | L = (Total likes + Total coins + Total favorite) / Total views * 100


```python
SumLikes = bibli['likes'].groupby(bibli['author']).sum()
SumLikes
```

```python
SumCoins = bibli['coins'].groupby(bibli['author']).sum()
SumCoins
```

```python
SumFavorite = bibli['favorite'].groupby(bibli['author']).sum()
SumFavorite
```

```python
l = (SumLikes+SumCoins+SumFavorite)/SumView*100
l = l.reset_index()
l.columns = ['author', 'L']
l
```

```python
# Using the pd.merge() function in-function connection, merge merge I, F, L horizontally, and assign a value to the IFL

ifa = pd.merge(i, f, on = 'author', how = 'inner')
ifl = pd.merge(ifa, l, on = 'author', how = 'inner')
ifl 
```

```python
# Name the IFL's column name "author," "I," "F," "L"

ifl = ifl[["author","I","F","L"]]
```

### Step 3 marks the hierarchy

#### 3.1 Drop Duplicates

```python
# Use the drop_duplicates() function to remove duplicate values from the ifl

ifl = ifl.drop_duplicates()
```

#### 3.2 Seperate the bins

```python
# Use the qcut() function
# The data bins of "I", "F" and "L" are divided into 5 groups, with the interval marker named 1-5
# And assigned to the ifl['I-SCORE'], the ifl['F-SCORE'] and the ifl['L-SCORE'] respectively.

ifl['I-SCORE'] = pd.qcut(ifl['I'], q = 5, labels = [1, 2, 3, 4, 5])
ifl['F-SCORE'] = pd.qcut(ifl['F'], q = 5, labels = [1, 2, 3, 4, 5])
ifl['L-SCORE'] = pd.qcut(ifl['L'], q = 5, labels = [1, 2, 3, 4, 5])

ifl
```

#### 3.3 Convert to 0/1 labels

```python
# Define a function: rfmTrans, which returns 1 when parameter x>3, else return 0;

def rfmTrans(x):
    if x > 3:
        return 1
    else:
        return 0
```

```python
# Apply the rfmTrans() function to the ifl model

ifl['I-SCORE'] = ifl['I-SCORE'].apply(rfmTrans)
ifl['F-SCORE'] = ifl['F-SCORE'].apply(rfmTrans)
ifl['L-SCORE'] = ifl['L-SCORE'].apply(rfmTrans)
ifl
```

```python
# Use the astype() function to convert the I_score, F_score, L_score, and columns into string formats
# And stitch the strings together with a "plus" and assign them to the IFL .

ifl["mark"] = ifl["I-SCORE"].astype(str) + ifl["F-SCORE"].astype(str) + ifl["L-SCORE"].astype(str)
```

#### 3.4 Conversion of ingredient layer results

```python
# Define a function, rfmType, that converts values such as "111" and "110" into corresponding user hierarchies

def rfmType(x):
    if x == "111":
        return "High Quality Author"
    elif x == "101":
        return "High Quality Lazy Author"
    elif x == "011":
        return "High Quality Content and High Depth Author"
    elif x == "001":
        return "High Quality Content High Depth and Lazy Author"
    elif x == "110":
        return "Down to Earth and Active Author"
    elif x == "100":
        return "Down to Earth Author"
    elif x == "010":
        return "Active Author"
    else:
        return "Growing Author"
```

```python
# For the "mark" column, use the apple() function and the rfmType() function, assign the value to the ifl("up_type")

ifl["author_type"] = ifl["mark"].apply(rfmType)
```

#### For 3.5 statistics

```python
# Use groupby() and count() functions to count the number of authors for each author type

df_type = ifl["author_type"].groupby(ifl["author_type"]).count()
print(df_type)
```
